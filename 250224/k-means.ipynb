```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Yapay veri seti oluşturuyoruz
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# İlk 5 veriye bakmak
print(X[:5])
```

```python
# K-means modelini oluşturuyoruz (4 küme ile)
kmeans = KMeans(n_clusters=4, random_state=42)

# Modeli veriye uyguluyoruz
kmeans.fit(X)

# Kümeleri ve her bir veriye ait küme etiketlerini alıyoruz
y_kmeans = kmeans.predict(X)

# Küme merkezlerini alıyoruz
centers = kmeans.cluster_centers_

# Sonuçları yazdırmak
print("Küme merkezleri:\n", centers)
print("Veri kümesindeki her örneğin kümesi:\n", y_kmeans)
```

```python
# Veriyi ve kümeleri görselleştiriyoruz
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Küme merkezlerini çiziyoruz
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label="Küme Merkezleri")

plt.title('K-means Kümeleme Sonuçları')
plt.xlabel('Özellik 1')
plt.ylabel('Özellik 2')
plt.legend()
plt.show()
```
